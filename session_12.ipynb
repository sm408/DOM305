```python
from collections import defaultdict
import xml.etree.ElementTree as ET
```

```python
#reading the foodmenu file

xmlfile_plant = "xmlfiles/plant_catalog.xml"

tree_plant = ET.parse(xmlfile_plant)
print(tree_plant)

#get root element
root_plant = tree_plant.getroot()
print(root_plant)

for item in root_plant.iter():
        print(item.tag)
```

```python
print(root_plant.findall('./PLANT/'))
# ./ is known as xpath and gives the address of the dat in memory
```

```python
for item in root_plant.findall('./PLANT/'):
        print(item.tag,item.text)
```

```python
#add all tags in a dictionary using a function

def extract_data_xml(f1, arg1):
        tree = ET.parse(f1)

        #get root element
        root = tree.getroot()


        dict_xml_plant = defaultdict(list)

        for item in root.findall('./'+str(arg1)+'/'):
                dict_xml_plant[str(item.tag)].append(str(item.text))

        return dict_xml_plant

file_name = "xmlfiles/plant_catalog.xml"
arguement1 = "PLANT"
data_xml = extract_data_xml(file_name, arguement1)
print(data_xml)
```

```python
#dump xml data in a new json file

import json

jsonoutfile = "xmlfiles/plant_catalog.json"
with open(jsonoutfile,"w") as write_file:
    json.dump(data_xml, write_file)
```
