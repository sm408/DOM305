```python
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
import urllib.request
import time
import random, os, glob,json
```

```python
user_agent = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
headers = { 'User-Agent' : user_agent }
```

```python
#Extract the url links from previously downloaded business standard search pages

destdir = 'spring23_DOM305_webscraping/'
set1 = 'BuzStd_news_items/'
pathd = destdir+set1

if not os.path.exists(pathd) :
    os.makedirs(pathd)


def extract_newsitem_business_standard(souredir) :

    htmlfiles = glob.glob(souredir+'5-1-2020.html')

    for newsfiles in htmlfiles :

        # open the files
        file1 = open(newsfiles, 'r')

        #implement the BeautifulSoup library to read the html page
        soup1 = BeautifulSoup(file1, "lxml") #lxml is library xml
        ## you can change the "lxml" to "html.parser"

        souplist = soup1('ul', {"class":"listing"}) #this is the syntax
        #h2 means heading 2

        print(souplist[0])

        num_elements_list = len(souplist[0]('h2'))



        #check through the list of header 2 by looping 
        for ind in range(0, int(num_elements_list),1) :
            ## 'a' is for anchor and 'href' is the hyperlink
            business_standard_webpages = souplist[0]('h2')[int(ind)]('a')[0]['href']
            
            print(business_standard_webpages)
            urllink = "https://www.business-standard.com/"+str(business_standard_webpages.encode("utf-8"))

            pageinfo = urllink.split('/')[-1]
            print(urllink, pageinfo)

            # Download the urls
            base_url = urllib.request.Request(urllink, headers=headers)
            urlopen = urllib.request.urlopen(base_url)
            url_read = urlopen.read()
            
            filename = pathd+str(pageinfo) ## filename with the destination directory
           
            ### save the file
            file_out = open(filename,"w")
            print(url_read, file=file_out)
    
            #waiting time
            time.sleep(random.randint(3,5))

sourcedir = 'spring23_DOM305_webscraping/BS_set_3/'
extract_newsitem_business_standard(sourcedir)
```
